```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Montando o Drive
from google.colab import drive
drive.mount('/content/drive')

# Caminho do diretório escolhido
path = "/content/drive/MyDrive/DataScience"
```

```python
np.random.seed(42)
mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, 1000)

ponto_especifico = 65

z_score = (ponto_especifico - mean) / std_dev

# Gráfico
x = np.linspace(min(data), max(data), 1000)
y = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Distribuição Normal', color='blue')
plt.fill_between(x, y, color='lightblue', alpha=0.4)

# Ponto específico ao gráfico
plt.scatter(ponto_especifico, norm.pdf(ponto_especifico, mean, std_dev),
            color='red', label=f'Ponto ({ponto_especifico})\nZ-Score: {z_score:.2f}')
plt.vlines(ponto_especifico, ymin=0, ymax=norm.pdf(ponto_especifico, mean, std_dev),
           color='red', linestyle='dashed')


plt.title('Distribuição Normal com Ponto e Z-Score', fontsize=16)
plt.xlabel('Valor', fontsize=14)
plt.ylabel('Densidade', fontsize=14)
plt.legend(fontsize=12)
plt.grid(alpha=0.3)
plt.show()
```
